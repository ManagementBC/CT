```python
import time
import requests
import gradio as gr
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, T5Tokenizer, T5ForConditionalGeneration
import openai


# Initialize the tokenizer and model for Gemini (adjust paths accordingly)
gemini_tokenizer = AutoTokenizer.from_pretrained('/content/fine_tuned_gemini/content/fine_tuned_gemini')
gemini_model = AutoModelForSeq2SeqLM.from_pretrained('/content/fine_tuned_gemini/content/fine_tuned_gemini')
t5_tokenizer = T5Tokenizer.from_pretrained("/content/fine_tuned_t5/content/t5_finetuned_clinical_trial")
t5_model = T5ForConditionalGeneration.from_pretrained("/content/fine_tuned_t5/content/t5_finetuned_clinical_trial")

# Set OpenAI API key for GPT-4

openai.api_key = ""

# Function to fetch patient data from IPFS
def fetch_patient_data_from_ipfs(patient_id, max_retries=3):
    try:
        # Fetch IPFS hash from the smart contract
        ipfs_hash = contract_instance.functions.patients(patient_id).call()[3]
        print(f"Fetched IPFS hash for patient ID {patient_id}: {ipfs_hash}")
    except Exception as e:
        print(f"Error fetching IPFS hash from smart contract: {str(e)}")
        return None

    # Construct the Pinata URL (or use another gateway)
    url = f"https://violet-big-lemur-828.mypinata.cloud/ipfs/{ipfs_hash}"
    print(f"Fetching data from URL: {url}")

    retries = 0
    while retries < max_retries:
        try:
            response = requests.get(url)

            if response.status_code == 200:
                # Parse the JSON data from the response
                try:
                    patient_data = response.json()
                except ValueError:
                    print(f"Error: Expected JSON but received different format for patient ID {patient_id}.")
                    return None

                print("Data successfully fetched from IPFS.")
                return patient_data

            elif response.status_code == 429:
                print(f"Rate limit hit. Retrying in {2 ** retries} seconds...")
                time.sleep(2 ** retries)
                retries += 1
            else:
                print(f"Failed to fetch patient data from IPFS. Status Code: {response.status_code}.")
                return None
        except requests.exceptions.RequestException as e:
            print(f"Error during request to Pinata: {str(e)}")
            return None

    print("Max retries reached for fetching patient data from IPFS.")
    return None


# Function to evaluate patient data using the Gemini model
def evaluate_with_gemini(patient_id, trial_criteria):
    # Fetch the patient data from IPFS
    patient_data = fetch_patient_data_from_ipfs(patient_id)
    if not patient_data:
        return f"Not Eligible, Patient ID: {patient_id}", patient_data

    # Safely access the patient data fields using .get() to avoid KeyError
    age = patient_data.get('Age', 'N/A')  # Default to 'N/A' if Age is missing
    conditions = patient_data.get('Conditions', [])
    biomarkers = patient_data.get('Biomarkers', {})
    hba1c = biomarkers.get('HbA1c', 'N/A')  # Default to 'N/A' if HbA1c is missing
    blood_glucose = biomarkers.get('Blood Glucose', 'N/A')  # Default to 'N/A' if Blood Glucose is missing

    # Structure the patient data for input into the model
    patient_data_str = (f"Patient ID: {patient_data.get('Patient ID', 'Unknown')}, "
                        f"Age: {age}, Conditions: {conditions}, "
                        f"Biomarkers: HbA1c: {hba1c}, Blood Glucose: {blood_glucose}")

    # Combine trial criteria and patient data into a single input string
    input_text = f"Trial Criteria: {trial_criteria}\nPatient Data: {patient_data_str}"

    # Print the combined input for debugging purposes
    print(f"Gemini Input for Patient ID {patient_id}: {input_text}")

    # Tokenize the input and generate the model's output
    input_ids = gemini_tokenizer(input_text, return_tensors="pt", truncation=True, padding="max_length", max_length=512).input_ids
    output_ids = gemini_model.generate(input_ids, max_length=50, num_beams=5, temperature=0.7, do_sample=True, early_stopping=True)

    # Decode the model's output (RAW response)
    raw_response = gemini_tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Print the raw response for debugging
    print(f"Raw response from Gemini model: {raw_response}")

    # Return the raw response directly for Gradio
    return raw_response, patient_data





def evaluate_with_t5(patient_id, trial_criteria):
    patient_data = fetch_patient_data_from_ipfs(patient_id)
    if not patient_data:
        return f"Not Eligible, Patient ID: {patient_id}", patient_data

    # Construct patient data string
    patient_data_str = (f"Patient ID: {patient_data.get('Patient ID', 'Unknown')}, "
                        f"Age: {patient_data.get('Age', 'N/A')}, "
                        f"Conditions: {patient_data.get('Conditions', [])}, "
                        f"Biomarkers: HbA1c: {patient_data.get('Biomarkers', {}).get('HbA1c', 'N/A')}, "
                        f"Blood Glucose: {patient_data.get('Biomarkers', {}).get('Blood Glucose', 'N/A')}")

    input_text = f"Trial Criteria: {trial_criteria}\nPatient Data: {patient_data_str}"
    print(f"T5 Input for Patient ID {patient_id}: {input_text}")

    input_ids = t5_tokenizer(input_text, return_tensors="pt", truncation=True, padding="max_length", max_length=512).input_ids
    output_ids = t5_model.generate(input_ids, max_length=50, num_beams=5, temperature=0.5, do_sample=False, early_stopping=True)

    raw_response = t5_tokenizer.decode(output_ids[0], skip_special_tokens=True)
    print(f"Raw response from T5 model: {raw_response}")

    return raw_response, patient_data

# Function to evaluate patient data using the GPT-4 model
def evaluate_with_gpt4(patient_id, trial_criteria):
    patient_data = fetch_patient_data_from_ipfs(patient_id)
    if not patient_data:
        return f"Not Eligible, Patient ID: {patient_id}", patient_data

    # Construct messages for GPT-4 input
    messages = [
        {"role": "system", "content": "You are a clinical trial eligibility checker."},
        {"role": "user", "content": f"Trial Criteria: {trial_criteria}\nPatient Data: Patient ID: {patient_data.get('Patient ID', 'Unknown')}, Age: {patient_data.get('Age', 'N/A')}, Conditions: {patient_data.get('Conditions', [])}, Biomarkers: HbA1c: {patient_data.get('Biomarkers', {}).get('HbA1c', 'N/A')}, Blood Glucose: {patient_data.get('Biomarkers', {}).get('Blood Glucose', 'N/A')}"}
    ]
    print(f"GPT-4 Input for Patient ID {patient_id}: {messages}")

    response = openai.ChatCompletion.create(
        model="ft:gpt-4o-mini-2024-07-18:personal:clincialtrailswithpaitents3:AKkOcg4l",
        messages=messages,
        max_tokens=150,
        temperature=0.5,  # Lower temperature for consistency
        top_p=1,
        n=1
    )

    raw_response = response.choices[0].message['content']
    print(f"Raw response from GPT-4 model: {raw_response}")

    return raw_response, patient_data

# Updated gradio_interface function to handle the selection of Gemini, T5, or GPT-4 models
def gradio_interface(trial_criteria, patient_ids, trial_id, model_choice):
    print(f"Selected Model: {model_choice}")
    print(f"Trial Criteria: {trial_criteria}")
    print(f"Trial ID: {trial_id}")

    # Ensure trial_id is an integer
    trial_id = int(trial_id)

    # Split patient_ids by comma and remove any extra spaces
    patient_ids_list = [pid.strip() for pid in patient_ids.split(",")]

    results = []
    for patient_id in patient_ids_list:
        print(f"Processing Patient ID: {patient_id}")

        # Convert the patient ID to an integer
        patient_id = int(patient_id)

        if model_choice == "Gemini":
            raw_response, patient_data = evaluate_with_gemini(patient_id, trial_criteria)
        elif model_choice == "T5":
            raw_response, patient_data = evaluate_with_t5(patient_id, trial_criteria)
        elif model_choice == "GPT-4":
            raw_response, patient_data = evaluate_with_gpt4(patient_id, trial_criteria)
        else:
            raw_response = "Model not yet implemented."
            patient_data = None

        # Keep the raw response intact
        raw_response_cleaned = raw_response.strip()
        raw_response_cleaned = raw_response_cleaned.replace("Condition", "").strip()
        raw_response_cleaned = raw_response_cleaned.replace(", Age", "").strip()

        # Determine eligibility by looking for "Not Eligible" or "Eligible"
        if "not eligible" in raw_response_cleaned.lower():
            is_eligible = False
        elif "eligible" in raw_response_cleaned.lower():
            is_eligible = True
        else:
            is_eligible = None

        if is_eligible is not None:
            tx_receipt = trigger_smart_contract(patient_id, is_eligible, model_choice, trial_id)

        results.append(f"Response: {raw_response_cleaned}")

    return "\n".join(results)

# LLM-specific credentials
llm_credentials = {
    "Gemini": {
        "account_address": "0x.........",
        "private_key": ""
    },
    "T5": {
        "account_address": "0x..........",
        "private_key": ""
    },
    "GPT-4": {
        "account_address": "0x..........",
        "private_key": ""
    }
}


    # Modify the trigger_smart_contract function to accept trial_id as input
def trigger_smart_contract(patient_id, is_eligible, model, trial_id):
    try:


        # Dynamically select credentials based on the model
        sender_address = llm_credentials[model]["account_address"]
        private_key = llm_credentials[model]["private_key"]

        # Set up the transaction
        nonce = w3.eth.get_transaction_count(sender_address)
        transaction = contract_instance.functions.evaluateEligibility(
            trial_id,   # Use trial_id provided by the user (uint256)
            patient_id,  # patientId (uint256)
            is_eligible,  # isEligible (True or False, bool)
            model       # Model used (string)
        ).build_transaction({
            'from': sender_address,
            'nonce': nonce,
            'gas': 2000000,  # Adjust gas limit
            'gasPrice': Web3.to_wei('20', 'gwei')  # Adjust gas price
        })

        # Sign the transaction with the private key
        signed_txn = w3.eth.account.sign_transaction(transaction, private_key=private_key)

        # Send the signed transaction
        tx_hash = w3.eth.send_raw_transaction(signed_txn.raw_transaction)

        # Wait for the transaction receipt
        receipt = w3.eth.wait_for_transaction_receipt(tx_hash)

        print(f"Transaction successful with hash: {tx_hash.hex()}")
        return receipt

    except Exception as e:
        print(f"Error calling smart contract function: {str(e)}")
        return str(e)

# Gradio interface definition remains the same, with the models Gemini, T5, and GPT-4 as choices
interface = gr.Interface(
    fn=gradio_interface,
    inputs=[
        gr.Textbox(lines=5, label="Enter Clinical Trial Criteria"),
        gr.Textbox(label="Enter Patient IDs (comma-separated)"),
        gr.Textbox(label="Enter Trial ID"),
        gr.Radio(choices=["Gemini", "T5", "GPT-4"], label="Select LLM Model"),
    ],
    outputs="text",
    title="Clinical Trial Eligibility Checker",
    description="Provide clinical trial criteria, patient IDs (comma-separated), trial ID, and select a model to evaluate eligibility."
)

# Launch the Gradio interface
interface.launch(debug=True)
```
